# 2016 v. 2017 Youth and Opportunity Youth by Age Group Visualizations

### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Total Opportunity Youth, 2016 v. 2017 Chart

```python
n_groups = 3
data_2016 = (6, 17, 18) 
data_2017 = (6, 15, 14) 

fig, ax = plt.subplots(figsize= (6,3))
index = np.arange(n_groups)
bar_width = 0.40
opacity = 0.7

bar2016= plt.bar(index, data_2016, bar_width,
alpha=opacity,
color='c',
label='2016')

bar2017= plt.bar(index + bar_width, data_2017, bar_width,
alpha=opacity,
color='b',
label='2017')

plt.xlabel('Age Groups (years old)', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.title('Total Opportunity Youth', fontsize=14)
plt.xticks(index + bar_width/2, ('16-18', '19-21', '22-24'), fontsize=10)
ax.set_ylim([0,20])
plt.legend()

plt.tight_layout()
plt.show()
plt.savefig('OY_total_20162017.png',dpi=300, bbox_inches='tight') 
```
