<h1>Importing Libraries</h1>

```python
import folium
import pandas as pd
import math
import branca as cm
from folium.plugins import FloatImage
```

<h1>Creating Dictionaries</h1>
Created a dictionary that has the Puma numbers and the OY population as well as a dictionary that includes the lattitude, longitute, area and Total population. 

```python
#Initial dictionary used for just the population
pop_dic = {'11604': 848.0,
 '11610': 1846.0,
 '11611': 1951,
 '11612': 1888.0,
 '11613': 1862.0,
 '11614': 1426.0,
 '11615': 1029.0}
```

```python
# additional imformation. 
#puma[name,OYpop,lat,lon,area in miles, total population]
reg = {'11612': ['Federal Way,Des moines Cities & Vashon island',  pop_dic['11612'], 47.3223, -122.3126, 110, 134000],
 '11604': ['South East Seattle',  pop_dic['11604'], 47.5923, -122.3022, 30.5, 84000],
 '11615': ['Maple Valley, Covingtion and Enumclaw Cities',  pop_dic['11615'], 47.3903, -122.0454, 18, 47000],
 '11614': ['Auburn City & Lakeland',  pop_dic['11614'], 47.3073, -122.2285, 35, 192000],
 '11613': ['Kent City',  pop_dic['11613'], 47.3809, -122.2348, 34.43, 130000],
 '11611': ['Burien, SeaTac, Tukwita Cities',  pop_dic['11611'], 47.4668, -122.3405, 31, 101000],
 '11610': ['Renton City, Fairwood, Bryn Mawr and Skyway', pop_dic['11610'], 47.4797, -122.2079, 31, 141]}
```

Assigned values to a specific variable in case I wanted to use it for plotting purposes.

```python
name = [i[0] for i in reg.values()]
population = [i[1] for i in reg.values()]
lat = [i[2] for i in reg.values()]
lon = [i[3] for i in reg.values()]
area = [i[4] for i in reg.values()]
```

<h1>Plotting the Map</h1>
Used Folium and the information from the 2017 pums data to give a visualization showing the regions and how many OY there are. 

```python
m = folium.Map(location=[47.45,-122.3321], zoom_start=11)

#loop through each lat,long and population. 
#
for i in range(0,len(reg.values())):
   folium.CircleMarker(
      location=[lat[i], lon[i]],
      popup= name[i] + ', \nPopulation:' + str(population[i]),
      radius=population[i]/20,
      color='blue',
      fill=True,
      fill_color='blue'
   ).add_to(m)

#creates a legend in the upper left portion. 
colormap = cm.colormap.LinearColormap(colors=['blue','blue','blue','blue','blue'], index=[0,1000,2000,3000,4000],vmin=0,vmax=4000);
colormap.caption = 'POPULATION OF OPPORTUNITY YOUTH';
colormap.add_to(m);
m
```

```python
#saves the map
m.save('opportunity_youth_map.png')
```
